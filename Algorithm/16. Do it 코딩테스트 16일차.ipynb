## Q54

```python
from collections import deque
import sys
input = sys.stdin.readline
del input

N = int(input())
A = [[] for _ in range(N+1)]
Time = [0] * (N+1)
Degree = [0] * (N+1)
Answer = [0] * (N+1)

for i in range(1, N+1):
    temp = list(map(int, input().split()))
    Time[i] = temp[0]
    for j in temp[1:]:
        if j == -1 : break
        A[j].append(i)
        Degree[i] += 1

q = deque()

for i in range(1, N+1):
    if Degree[i] == 0 : q.append(i)

while q:
    now = q.popleft()
    for next in A[now]:
        Degree[next] -= 1
        Answer[next] = max(Answer[next] , Answer[now]+Time[now])
        if Degree[next] == 0 : q.append(next)
            
for i in range(1, N+1):
    Answer[i] += Time[i]
    print(Answer[i])
```

## Q55

```python
from collections import deque
import sys
input = sys.stdin.readline
del input

N = int(input())
M = int(input())
A = [[] for _ in range(N+1)]
reverseA = [[] for _ in range(N+1)]
Degree = [0] * (N+1)
Result = [0] * (N+1)

for i in range(M):
    s, e, v = map(int, input().split())
    A[s].append((e,v))
    reverseA[e].append((s,v))
    Degree[e] += 1
    
Start, End = map(int, input().split())

q = deque()
q.append(Start)

while q:
    now = q.popleft()
    for next in A[now]:
        Degree[next[0]] -= 1
        Result[next[0]] = max(Result[next[0]], Result[now] + next[1])
        if Degree[next[0]] == 0 :
            q.append(next[0])

ResultCount = 0
visited = [False] * (N+1)
q.clear()
q.append(End)
visited[End] = True

while q:
    now = q.popleft()
    for next in reverseA[now]:
        if Result[next[0]] + next[1] == Result[now]:
            ResultCount += 1
            if not visited[next[0]]:
                visited[next[0]] = True
                q.append(next[0])
                
print(Result[End])
print(ResultCount)
```
